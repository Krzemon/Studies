# Three channels and colors

 > ways of organising data in `numpy` arrays   

The images are represented in Python as number arrays. For a single-chanel image, the array is a two-dimensional grid with number of rows (n_rows) on number of columns (n_cols).

Multi-channel images are represented as arrays of dimensions:
 - number of channels (bands), number of rows, number of columns - `b x r x c`
 - number of rows, number of columns, number of channels (bands) - `r x c x b`

Function for ploting image + histogram - to make life easier 

```python
def plotimhist(image):
    
    plt.figure(figsize = (12,6))
    ax1 = plt.subplot(1,2,1)
    ax1.imshow(image, vmin=0, vmax=255)
    ax2 = plt.subplot(1,2,2)
    ax2.hist(image[:,:,0].flatten(), bins=255, color = 'red', alpha = 0.5);
    ax2.hist(image[:,:,1].flatten(), bins=255, color = 'green', alpha = 0.5);
    ax2.hist(image[:,:,2].flatten(), bins=255, color = 'blue', alpha = 0.5);
```

```python
def plotRGB(image):

    plt.figure(figsize = (18,6))
    ax1 = plt.subplot(1,3,1)
    ax1.imshow(image[:,:,0], cmap=plt.cm.Greys_r, vmin=0, vmax=255)
    ax1.set_title('RED')
    ax2 = plt.subplot(1,3,2)
    ax2.imshow(image[:,:,1], cmap=plt.cm.Greys_r, vmin=0, vmax=255)
    ax2.set_title('GREEN')
    ax3 = plt.subplot(1,3,3)
    ax3.imshow(image[:,:,2], cmap=plt.cm.Greys_r, vmin=0, vmax=255)
    ax3.set_title('BLUE')
```

# Color conversions
## RGB to Gray
The value of each grayscale pixel is calculated as the weighted sum of the corresponding red, green and blue pixels as:
> Y = 0.2125 R + 0.7154 G + 0.0721 B

> skimage method: skimage.color.rgb2gray

## RGB to HSV
RGB to [HSV](https://en.wikipedia.org/wiki/HSL_and_HSV) (Hue, Saturation, Value)
> skimage method: skimage.color.rgb2hsv
