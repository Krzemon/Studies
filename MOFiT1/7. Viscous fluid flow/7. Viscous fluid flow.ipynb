```python
dz = 0.01
X, Y = 250, 80
y_1, y_2 = -0.4, 0.4
Q = -1
mi = 1

xx = np.arange(-100, 151, 1)
yy = np.arange(-40, 41, 1)

ik = 5
jk = 10
X1 = -100
X2 = 150
Y1 = -40
Y2 = 40

Q1 = Q
Q2 = -10
Q3 = -100
Q4 = -200
Q5 = -400
```

```python
def brzeg(A, B):
    for i in range(X + 1):
        for j in range(Y + 1):
            if xx[i] == X2 or yy[j] == Y2 or xx[i] == X1 or yy[j] == Y1:
                A[i, j] = Q / (2 * mi) * (math.pow(y[j], 3) / 3 + y_1 * y_2 * y[j])
                B[i, j] = Q / (2 * mi) * (2 * y[j])

def u≈õrednij(A, B):
    for i in range(1, X):
        for j in range(1, Y):
            A[i, j] = (A[i + 1, j] + A[i - 1, j] + A[i, j - 1] + A[i, j + 1] - B[i, j] * dz * dz) / 4
            B[i, j] = (B[i + 1, j] + B[i - 1, j] + B[i, j - 1] + B[i, j + 1]) / 4 - ((A[i, j + 1] - A[i, j - 1]) * (B[i + 1, j] - B[i - 1, j]) - (A[i + 1, j] - A[i - 1, j]) * (B[i, j + 1] - B[i, j - 1])) / 16
            
def teoria(A, B): 
    for j in range(Y + 1):
        A[0, j] = Q / (2 * mi) * (math.pow(y[j], 3) / 3 + y_1 * y_2 * y[j])
        B[0, j] = Q / (2 * mi) * (2 * y[j])
        A[1, j] = Q / (2 * mi) * (math.pow(y[j], 3) / 3 + y_1 * y_2 * y[j])
        B[1, j] = Q / (2 * mi) * (2 * y[j])
```

```python
psi = np.zeros((X + 1, Y + 1))
ksi = np.zeros((X + 1, Y + 1))

brzeg(psi, ksi)
temp_psi = 0
temp_ksi = 0

for i in range(4800):
    u≈õrednij(psi, ksi)
    print(i)
    if abs(temp_psi - psi[50, 40]) < 0.0000001 and abs(temp_ksi - ksi[50, 40]) < 0.0000001 and i > 100:
        break
    temp_psi = psi[50, 40]
    temp_ksi = ksi[50, 40]
    
teory_psi = np.zeros((2, 81)) 
teory_ksi = np.zeros((2, 81))
teoria(teory_psi, teory_ksi)
```

```python
def brzegZastawka(A, B):
    for i in range(X+1):
        for j in range(Y+1):
            
            if yy[j] == Y1 and (xx[i] < -ik or xx[i] > ik):
                A[i,j] = Q/(2*mi) * (math.pow(y_1,3)/3 + y_1*y_2*y_1)

            if xx[i] == -ik and yy[j] <= jk:
                A[i,j] = Q/(2*mi) * (math.pow(y_1,3)/3 + y_1*y_2*y_1)

            if xx[i] == ik and yy[j] <= jk:
                A[i,j] = Q/(2*mi) * (math.pow(y_1,3)/3 + y_1*y_2*y_1)

            if -5 < xx[i] < ik and yy[j] == jk:
                A[i,j] = Q/(2*mi) * (math.pow(y_1,3)/3 + y_1*y_2*y_1)

            if xx[i] == X1 or xx[i] == X2:
                A[i,j] = Q/(2*mi) * (math.pow(y[j],3)/3 + y_1*y_2*y[j])
                
            if yy[j] == Y1:
                if -ik > xx[i] or xx[i] > ik:
                    B[i,j] = 2*(A[i,j+1]-A[i,j])/(dz*dz)

            if yy[j] == Y2:
                A[i,j] = Q/(2*mi) * (math.pow(y[j],3)/3 + y_1*y_2*y[j])
                B[i,j] = 2*(A[i,j-1] - A[i,j])/(dz*dz)

            if xx[i] == -ik and yy[j] < jk:
                B[i,j] = 2*(A[i-1,j] - A[i,j])/(dz*dz)

            if xx[i] == ik and yy[j] < jk:
                B[i,j] = 2*(A[i+1,j] - A[i,j])/(dz*dz)

            if 5 < xx[i] < ik and yy[j] == jk:
                B[i,j] = 2*(A[i,j+1] - A[i,j])/(dz*dz)

            if xx[i] == -ik and yy[j] == jk:
                B[i,j] = (A[i,j+1] - A[i,j])/(dz*dz) + (A[i-1,j] - A[i,j])/(dz*dz)

            if xx[i] == ik and yy[j] == jk:
                B[i,j] = (A[i,j+1] - A[i,j])/(dz*dz) + (A[i+1,j] - A[i,j])/(dz*dz)

def u≈õrednijZastawka(A, B):
    for i in range(1,X):
        for j in range(1,Y):
            if not (-ik <= xx[i] <= ik and yy[j] <= jk):
                A[i,j] = (A[i+1,j] + A[i-1,j] + A[i,j-1] + A[i,j+1] - B[i,j]*dz*dz)/4
                B[i,j] = (B[i+1,j] + B[i-1,j] + B[i,j-1] + B[i,j+1])/4 -((A[i,j+1] - A[i,j-1]) * (B[i+1,j] - B[i-1,j]) - (A[i+1,j]-A[i-1,j])*(B[i,j+1] - B[i,j-1]))/16

def wyzerujPrzeszkode(A, B):
    for i in range(0,X):
        for j in range(0,Y):
            if -ik < xx[i] < ik and yy[j] < jk:
                A[i,j] = 0
                B[i,j] = 0

def pochodna(macierz):
    A = np.zeros(np.shape(macierz))
    B = np.zeros(np.shape(macierz))

    for i in range(len(macierz)-1):
        for j in range(len(macierz[0])-1):
            if not(-ik-1 < xx[i] and ik > xx[i] and yy[j] < jk): 
                A[i,j] = (macierz[i+1,j] - macierz[i,j])/dz
            if not(-ik < xx[i] and ik > xx[i] and yy[j] < jk): 
                B[i,j] = (macierz[i,j+1] - macierz[i,j])/dz
    return A, B
```

```python
psi_z = np.copy(psi)
ksi_z = np.copy(ksi)
wyzerujPrzeszkode(psi_z, ksi_z)

for k in range(500):
    u≈õrednijZastawka(psi_z, ksi_z)
    brzegZastawka(psi_z, ksi_z)
    print(k)
```
