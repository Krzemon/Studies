```python
import numpy as np
from numpy import linalg as LA
import matplotlib.pyplot as plt
```

```python
n = 1000
m = 5
A = np.zeros((n,n))

for i in range(0,n):
    for j in range(0,n):
        temp = abs(i - j)
        if temp <= m:
            A[i][j] = 1 / (1 + temp)
        else:
            A[i][j] = 0
    
b = np.arange(n)
```

```python
r_a = np.zeros(n)
x_a = np.zeros(n)

temp_vec = np.zeros(n)
alpha_a = 0
NORM_X_A = [0]
NORM_R_A = [0]

k_a = 1 # pierwsza iteracja
r_a = b - A.dot(x_a)
NORM_R_A.append(LA.norm(r_a))

print(f"{k_a}: r_a_norm = {LA.norm(r_a)}, alpha_a = {alpha_a}, x_a_norm = {LA.norm(x_a)}")
alpha_a = np.inner(r_a,r_a) / np.inner(A.dot(r_a),r_a)
x_a += alpha_a * r_a
NORM_X_A.append(LA.norm(x_a))

while True:
    r_a = b - A.dot(x_a)
    NORM_R_A.append(LA.norm(r_a))
    alpha_a = np.inner(r_a,r_a) / np.inner(A.dot(r_a),r_a)
    x_a += alpha_a * r_a
    NORM_X_A.append(LA.norm(x_a))
    norm = LA.norm(r_a)
    k_a += 1
    print(f"{k_a}: r_a_norm = {norm}, alpha_a = {alpha_b}, x_a_norm = {LA.norm(x_a)}")

    if abs(norm) < 1e-6:
        break
```

```python
r_b = np.zeros(n)
x_b = np.ones(n)

temp_vec = np.zeros(n)
alpha_b = 0
NORM_X_B = [0]
NORM_R_B = [0]

k_b = 1 # pierwsza iteracja
r_b = b - A.dot(x_b)
NORM_R_B.append(LA.norm(r_b))

print(f"{k_b}: r_b_norm = {LA.norm(r_b)}, alpha_b = {alpha_b}, x_b_norm = {LA.norm(x_b)}")
alpha_b = np.inner(r_b,r_b) / np.inner(A.dot(r_b),r_b)
x_b += alpha_b * r_b
NORM_X_B.append(LA.norm(x_b))

while True:
    r_b = b - A.dot(x_b)
    NORM_R_B.append(LA.norm(r_b))
    alpha_b = np.inner(r_b,r_b) / np.inner(A.dot(r_b),r_b)
    x_b += alpha_b * r_b
    NORM_X_B.append(LA.norm(x_b))
    norm = LA.norm(r_b)
    k_b += 1
    print(f"{k_b}: r_b_norm = {norm}, alpha_b = {alpha_b}, x_b_norm = {LA.norm(x_b)}")

    if abs(norm) < 1e-6:
        break
```
