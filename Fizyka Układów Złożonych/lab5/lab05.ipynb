```python
def rand_spins(spins):
    i = 0
    while i < N:
        if random.random() <= 0.5:
            spins.append(-1)
        else:
            spins.append(1)
        i += 1
    return spins
```

```python
# dla B = 0
# zwraca energie calkowita ukladu
def Energy(tab):
    E = 0
    for i in range(0, N):
        E += J * tab[i] * tab [(i+1) % N]
    return E
```

```python
# losujemy wstepna konfiguracje spinow
spins = []
rand_spins(spins)

T = T_odw[0]
E = Energy(spins)
MC_values = []
En_values = []
MC = 10000

# glowna petla monte carlo
for _ in range(0, MC):

    for spin_id in range(0, N):
        if spins[(spin_id - 1) % N] * spins[(spin_id + 1) % N] < 0 :
            dE = 0
            spins[spin_id] *= -1 # odwrocenie znaku
        elif spins[spin_id] == -1 and spins[(spin_id - 1) % N] + spins[(spin_id + 1) % N] == 2:
            dE = 4
            if random.random() < np.exp(-dE*T):
                spins[spin_id] *= -1
                E+=dE
        elif spins[(spin_id - 1) % N] == spins[(spin_id + 1) % N] and -1*spins[spin_id] == spins[(spin_id + 1) % N]:
            dE = -4
            spins[spin_id] *= -1
            E += dE
            
    E_perspin = E / N
    En_values.append(E_perspin)
```

```python
plt.title("energia dla 1/T = 0.5")
plt.plot(En_values, 'r.', label=(r'$\lambda = 0.3$'))
plt.savefig("zad2_a.pdf")
# plt.xlim(1,10)
plt.show()
```
