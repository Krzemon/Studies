```python
import numpy as np
import matplotlib.pyplot as plt
```

# Zadanie 1

```python
r = 2

def logistic_map(x):
    return r * x * (1 - x)

x0_values = np.linspace(0.1, 0.9, 9)
```

```python
plt.figure(figsize=(10, 6))

for x0 in x0_values:
    x = x0
    trajectory = [x]
    for _ in range(50):
        x = logistic_map(x)
        trajectory.append(x)
    plt.plot(trajectory, label=f'x0={x0:.1f}')

plt.xlabel('Krok czasowy')
plt.ylabel('Gęstość populacji')
plt.title('Ewolucja populacji dla różnych x0 przy r=2')
plt.legend()
plt.grid(True)
plt.show()
```

### Analiza wyników:
Na wykresie zobaczymy trajektorie dla różnych wartości początkowych gęstości populacji. Spodziewamy się, że trajektorie zbiegną do punktu równowagi x*, który jest rozwiązaniem równania x* = rx*(1 - x*), które można obliczyć numerycznie lub analitycznie. Widzimy tylko jedną wartość graniczną, ponieważ trajektorie zbiegają do jednego punktu równowagi x* dla ustalonej wartości parametru r i różnych wartości początkowych gęstości populacji x0.


# Zadanie 2

```python
x0 = 0.5
r_values = [1, 2, 3, 3.5, 3.55, 3.6]
plt.figure(figsize=(10, 6))

for r in r_values:
    x = x0
    trajectory = [x]
    for _ in range(100):
        x = r * x * (1 - x)
        trajectory.append(x)
    plt.plot(trajectory, label=f'r={r}')

plt.xlabel('Krok czasowy')
plt.ylabel('Gęstość populacji')
plt.title('Ewolucja populacji dla różnych wartości r przy x0=0,5')
plt.legend()
plt.grid(True)
plt.show()
```

### Analiza wyników:

Na wykresie zobaczymy trajektorie dla różnych wartości parametru `r` przy ustalonej wartości początkowej gęstości populacji `x0 = 0.5`. Zauważymy różne zachowanie trajektorii w zależności od wartości parametru `r`.

- Dla wartości `r` poniżej 1, trajektoria będzie zbiegać do zera.
- Dla wartości `r` między 1 a 3, trajektoria będzie zbiegać do jednej wartości granicznej.
- Dla wartości `r` powyżej 3, trajektoria będzie oscylować między różnymi wartościami. Osiąga to, co nazywa się chaosem deterministycznym.
- Dla wartości `r` blisko 3,5, zobaczymy występowanie atraktorów dziwnych, co oznacza, że trajektoria będzie oscylować wokół wielu różnych wartości.

# Zadanie 3

```python
N = 10_000
r_tab = np.linspace(3.54, 3.55, 1000)

for r in r_tab:
    x_tab = [0.5]
    for i in range(1, N + 1):
        x = r * x_tab[-1] * (1 - x_tab[-1])
        if i >= 9000:  # Zbieranie wartości tylko dla ostatnich 1000 kroków
            x_tab.append(x)
    plt.plot([r] * len(set(x_tab)), list(set(x_tab)), 'k.', markersize=0.5)  # Rysowanie unikalnych wartości x dla danego r

plt.xlabel('Wartość parametru r')
plt.ylabel('Końcowa wartość populacji')
plt.title('Końcowa wartość populacji w zależności od r')
plt.grid(True)
plt.savefig("zad3_3.pdf")
plt.show()
```
