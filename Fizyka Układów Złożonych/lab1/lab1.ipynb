```python
import numpy as np
import matplotlib.pyplot as plt
```

# Zadanie 1

```python
def FullGraph(N : int):
    matrix = np.ones((N,N))
    np.fill_diagonal(matrix, 0)
    return matrix

N = 8
mat = FullGraph(N)
print(str(mat))
```

```python
bar_x = np.arange(N)
bar_height = np.zeros(N)
for i in range(len(mat)):
    bar_height[int(sum(mat[i]))] += 1
bar_height = bar_height / np.sum(bar_height)

plt.bar(bar_x, bar_height)
plt.xlim(0, 8)
plt.title(f'Full Graph with N = {N} corners')
plt.xlabel('Rank')
plt.ylabel('Occurence')
plt.show()
```

# Zadanie 2

```python
def SparseGraph(N : int, chance : float):
    graph = []
    for i in range(N):
        corner = []
        for j in range(N):
            if i == j:
                continue
            
            if np.random.rand(1) <= chance:
                corner.append(j)
        graph.append(corner)    
    return graph

N = 100
chance = 1/3
sparseGraph = SparseGraph(N, chance)
print(sparseGraph[0])
```

```python
bar_x = np.arange(N)
bar_height = np.zeros(N)
for i in range(len(sparseGraph)):
    bar_height[len(sparseGraph[i])] += 1
bar_height = bar_height / np.sum(bar_height)

plt.bar(bar_x, bar_height, width=1)
plt.title(f'Sparse Graph with N = {N} corners')
plt.xlabel('Rank')
plt.ylabel('Occurence (normalized)')
plt.show()
```
