```python
import matplotlib.pyplot as plt
import random
import numpy as np

N=60
T=201
alpha=[0.1,0.2,0.3,0.3,0.4,0.5,0.6,0.7,0.8,0.9]
b=0.2




def neighbour_sick(A,y,x):
    sick = 0
    if(A[y][ (x+1)%N ] ==1 ):
        sick+=1
    if (A[y][x-1] == 1):
        sick += 1
    if (A[ (y+1)%N ][x-1] == 1):
        sick += 1
    if (A[ (y+1)%N ][x] == 1):
        sick += 1
    if (A[ (y+1)%N ][ (x+1)%N ] == 1):
        sick += 1
    if (A[y-1][x-1] == 1):
        sick += 1
    if (A[y-1][ (x+1)%N ] == 1):
        sick += 1
    if (A[y-1][x] == 1):
        sick += 1


    return sick

S_all = []  # 0-zdrowy/turkusowy
I_all = []  # 1chory/zolty
R_all = []  # -1/ozdrwialy granatowy
for a in alpha:

    S = []  # 0-zdrowy/turkusowy
    I = []  # 1chory/zolty
    R = []  # -1/ozdrwialy granatowy
    tab = np.zeros((N, N))
    p = int(N / 2)
    tab[p][p] = 1

    for t in range(T):
        temp = np.zeros((N, N))

        for y in range(N):
            for x in range(N):


                # zdrowy--zarażanie się od sąsiada z Prawd a
                if(tab[y][x]==0):
                    for i in range(neighbour_sick(A=tab,y=y,x=x)):
                        U1=random.random()
                        if(U1<a):
                            temp[y][x] = 1


                # chory, może ozdrowieć z prawd b-
                if (tab[y][x] == 1):
                    U1 = random.random()
                    if (U1 < b):
                        temp[y][x] = -1
                    else:
                        temp[y][x] = 1

                if (tab[y][x] == -1):
                    temp[y][x] = -1

        S.append( list(tab.reshape(-1)).count(0) )
        I.append(list(tab.reshape(-1)).count(1))
        R.append(list(tab.reshape(-1)).count(-1))



        # if t in [0,1,2,5,10,50,100]:
        #     plt.imshow(tab,vmin=-1,vmax=1)
        #     plt.title("t="+str(t))
        #     plt.savefig("zad1_t="+str(t)+"_a="+str(a)+".jpg")
        #     plt.show()
        tab = temp

    # plt.plot(S)
    # plt.plot(I)
    # plt.plot(R)

    # plt.xlabel("t")
    # plt.ylabel("N")
    # plt.savefig("zad2_a="+str(a)+".jpg")
    # plt.show()
    S_all.append(S)
    I_all.append(I)
    R_all.append(R)

for r,a in zip(R_all,alpha):
    plt.plot(r,label='a='+str(a))
plt.legend()
plt.savefig("zad3_R.jpg")
plt.plot()
```
