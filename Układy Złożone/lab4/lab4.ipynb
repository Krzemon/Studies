```python
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.animation as animation
from matplotlib.pyplot import imshow, show 
import time
```

```python
N = 60
times = [0, 1, 2, 5, 10, 50, 100]
Time = np.arange(0,100) + 1
α1 = 0.1
α2 = 0.2
α3 = 0.3
β1 = 0.1

S = 0 # susceptible
I = 1 # infected
R = 2 # recovered

matrix = np.zeros((N, N))
matrix[int(N/2)][int(N/2)] = 1 
```

```python
def neighbour_α(M,α,i,j):
    if(M[i][(j+1) % N] == 1 ):
        if np.random.rand(1) < α and M[i][j] != 2:
            M[i][j] = 1
    if (M[i][j-1] == 1):
        if np.random.rand(1) < α and M[i][j] != 2:
            M[i][j] = 1
    if (M[(i+1) % N][j-1] == 1):
        if np.random.rand(1) < α and M[i][j] != 2:
            M[i][j] = 1
    if (M[(i+1) % N][j] == 1):
        if np.random.rand(1) < α and M[i][j] != 2:
            M[i][j] = 1
    if (M[(i+1) % N][(j+1) % N] == 1):
        if np.random.rand(1) < α and M[i][j] != 2:
            M[i][j] = 1
    if (M[i-1][j-1] == 1):
        if np.random.rand(1) < α and M[i][j] != 2:
            M[i][j] = 1
    if (M[i-1][(j+1) % N] == 1):
        if np.random.rand(1) < α and M[i][j] != 2:
            M[i][j] = 1
    if (M[i-1][j] == 1):
        if np.random.rand(1) < α and M[i][j] != 2:
            M[i][j] = 1
    return M
```

```python
def neighbour_β(M,β,i,j):
    if(M[i][(j+1) % N] == 1 ):
        if np.random.rand(1) < β and M[i][j] == 1:
            M[i][j] = 2
    if (M[i][j-1] == 1):
        if np.random.rand(1) < β and M[i][j] == 1:
            M[i][j] = 2
    if (M[(i+1) % N][j-1] == 1):
        if np.random.rand(1) < β and M[i][j] == 1:
            M[i][j] = 2
    if (M[(i+1) % N][j] == 1):
        if np.random.rand(1) < β and M[i][j] == 1:
            M[i][j] = 2
    if (M[(i+1) % N][(j+1) % N] == 1):
        if np.random.rand(1) < β and M[i][j] == 1:
            M[i][j] = 2
    if (M[i-1][j-1] == 1):
        if np.random.rand(1) < β and M[i][j] == 1:
            M[i][j] = 2
    if (M[i-1][(j+1) % N] == 1):
        if np.random.rand(1) < β and M[i][j] == 1:
            M[i][j] = 2
    if (M[i-1][j] == 1):
        if np.random.rand(1) < β and M[i][j] == 1:
            M[i][j] = 2
    return M
```

# α = 0.1

```python
# for t in Time:
def infecting(matrix, alfa, beta):
    temp = np.zeros((N, N))
    temp = matrix
    for i in range(N):
        for j in range(N):
            temp = neighbour_α(temp, alfa, i, j)
            temp = neighbour_β(temp, beta, i, j)
    return temp
```

```python
plt.imshow(matrix, origin="lower", cmap='gray', interpolation='nearest')
```

```python
for t in Time:
    if t in times:
        plt.imshow(matrix, origin="lower", cmap='gray', interpolation='nearest')
        plt.title(f't = {t}')
        plt.savefig(f'./zad1a/time_{t}.png')
        plt.xticks([])
        plt.yticks([])
        plt.show()
    matrix = infecting(matrix, α1, β1)
```
