```python
import numpy as np
import matplotlib.pyplot as plt
```

# Zadanie 1 

```python
r = 2
x0_values = np.linspace(0.1, 0.9, 9)

def logistic_map(r,x):
    return r * x * (1 - x)

plt.figure(figsize=(10,7))
steps = 50 
for x0 in x0_values:
    track = []
    x = x0
    track.append(x)
    for _ in range(0,steps):
        x = logistic_map(r,x)
        track.append(x)
    plt.plot(track, label=f'x0={x0:.1f}')

plt.xlabel('Krok czasowy')
plt.ylabel('Gsto populacji')
# plt.yscale('log')
plt.title('Ewolucja populacji dla rnych x0 przy r=2')
plt.legend()
plt.grid(True)
plt.show()
```
