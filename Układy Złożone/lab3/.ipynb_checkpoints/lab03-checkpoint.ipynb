```python
import numpy as np
import matplotlib.pyplot as plt
```

# Zadanie 1

```python
r = 2

def logistic_map(x):
    return r * x * (1 - x)

x0_values = np.linspace(0.1, 0.9, 9)
```

```python
plt.figure(figsize=(10, 6))

for x0 in x0_values:
    x = x0
    trajectory = [x]
    for _ in range(50):
        x = logistic_map(x)
        trajectory.append(x)
    plt.plot(trajectory, label=f'x0={x0:.1f}')

plt.xlabel('Krok czasowy')
plt.ylabel('Gsto populacji')
plt.title('Ewolucja populacji dla rnych x0 przy r=2')
plt.legend()
plt.grid(True)
plt.show()
```

### Analiza wynikw:
Na wykresie zobaczymy trajektorie dla rnych wartoci pocztkowych gstoci populacji. Spodziewamy si, e trajektorie zbiegn do punktu rwnowagi x*, ktry jest rozwizaniem rwnania x* = rx*(1 - x*), ktre mona obliczy numerycznie lub analitycznie. Widzimy tylko jedn warto graniczn, poniewa trajektorie zbiegaj do jednego punktu rwnowagi x* dla ustalonej wartoci parametru r i rnych wartoci pocztkowych gstoci populacji x0.


# Zadanie 2

```python
x0 = 0.5
r_values = [1, 2, 3, 3.5, 3.55, 3.6]
plt.figure(figsize=(10, 6))

for r in r_values:
    x = x0
    trajectory = [x]
    for _ in range(100):
        x = r * x * (1 - x)
        trajectory.append(x)
    plt.plot(trajectory, label=f'r={r}')

plt.xlabel('Krok czasowy')
plt.ylabel('Gsto populacji')
plt.title('Ewolucja populacji dla rnych wartoci r przy x0=0,5')
plt.legend()
plt.grid(True)
plt.show()
```

### Analiza wynikw:

Na wykresie zobaczymy trajektorie dla rnych wartoci parametru `r` przy ustalonej wartoci pocztkowej gstoci populacji `x0 = 0.5`. Zauwaymy rne zachowanie trajektorii w zalenoci od wartoci parametru `r`.

- Dla wartoci `r` poniej 1, trajektoria bdzie zbiega do zera.
- Dla wartoci `r` midzy 1 a 3, trajektoria bdzie zbiega do jednej wartoci granicznej.
- Dla wartoci `r` powyej 3, trajektoria bdzie oscylowa midzy rnymi wartociami. Osiga to, co nazywa si chaosem deterministycznym.
- Dla wartoci `r` blisko 3,5, zobaczymy wystpowanie atraktorw dziwnych, co oznacza, e trajektoria bdzie oscylowa wok wielu rnych wartoci.

# Zadanie 3

```python
N = 10_000
r_tab = np.linspace(3.54, 3.55, 1000)

for r in r_tab:
    x_tab = [0.5]
    for i in range(1, N + 1):
        x = r * x_tab[-1] * (1 - x_tab[-1])
        if i >= 9000:  # Zbieranie wartoci tylko dla ostatnich 1000 krokw
            x_tab.append(x)
    plt.plot([r] * len(set(x_tab)), list(set(x_tab)), 'k.', markersize=0.5)  # Rysowanie unikalnych wartoci x dla danego r

plt.xlabel('Warto parametru r')
plt.ylabel('Kocowa warto populacji')
plt.title('Kocowa warto populacji w zalenoci od r')
plt.grid(True)
plt.savefig("zad3_3.pdf")
plt.show()
```
