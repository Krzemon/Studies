```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.linalg import eigh
import pandas as pd
```

```python
##################
# Parametry pudła#
##################
N = 9
NN = 2*N
Siatka = np.zeros((N,N))

Δx = Δy = 2 / 0.05292 # atomowe
a = Δx * (N - 1) / 2
x_values = np.zeros(N)
y_values = np.zeros(N)

for n in range(0, N):
    x_values[n] = -a + Δx * n
    y_values[n] = -a + Δy * n 

########################
# meV
########################
hw_x = 80
hw_y = 200
E_h = 27211.6
m = 0.24
ħ = 1

w_x = hw_x / E_h
w_y = hw_y / E_h

#######################
# szerokosc gaussianu #
#######################
alpha_x = ħ / (m * w_x)
alpha_y = ħ / (m * w_y)
########################
```

```python
# baza w której rozwijana jest funkcja falowa
# Funkcje falowe dla potencjału harmonicznego

def φ(r, μ, α):
    exp_pow = - np.power(r - μ, 2) / (2 * α)
    return np.exp(exp_pow) / np.power(α * np.pi,1/4)
```

```python
num = 100

# zagęszczenie punktów siatki
Xa = np.linspace(-a, a, num)
Ya = np.linspace(-a, a, num)

psi_0 = np.zeros((num, num))
for i in range(num):
    for j in range(num):
        psi_0[i, j] = φ(Xa[i], x_values[0], alpha_x) * φ(Ya[j], y_values[0], alpha_y)

psi_8 = np.zeros((num, num))
for i in range(num):
    for j in range(num):
        psi_8[i, j] = φ(Xa[i], x_values[0], alpha_x) * φ(Ya[j], y_values[8], alpha_y)

psi_9 = np.zeros((num, num))
for i in range(num):
    for j in range(num):
        psi_9[i, j] = φ(Xa[i], x_values[1], alpha_x) * φ(Ya[j], y_values[0], alpha_y)
```

## Zadanie 2: Wypełnianie tablic elementów macierzy S i H

```python
a /= 0.05292 # powrot do jednostek atomowych

# Całka przekrywania 
def overlap_integral(x1, x2, y1, y2, alpha_x, alpha_y):
    return np.exp( -0.25 * np.power(x1 - x2, 2) / alpha_x - 0.25 * np.power(y1 - y2, 2) / alpha_y )

# Całka kinetyczna
def kinetic_integral(x1, x2, y1, y2, alpha_x, alpha_y):
    return -1 / (2*m) * (0.25 * (np.power(x1 - x2, 2) - 2*alpha_x) / np.power(alpha_x, 2) + 0.25 * (np.power(y1 - y2, 2) - 2*alpha_y) / np.power(alpha_y, 2)) * overlap_integral(x1, x2, y1, y2, alpha_x, alpha_y)

# Całka potencjalna
def potential_integral(x1, x2, y1, y2, omega_x, omega_y, alpha_x, alpha_y):
    return 1 / (2*m) * (0.25 * np.power(omega_x, 2) * (np.power(x1 + x2, 2) + 2*alpha_x) + 0.25 * np.power(omega_y, 2) * (np.power(y1 + y2, 2) + 2*alpha_y)) * overlap_integral(x1, x2, y1, y2, alpha_x, alpha_y)

# Wypełnienie macierzy S i H
def calculate_S_H_matrices(N, alpha_x, alpha_y, omega_x, omega_y):
    S = np.zeros((N, N))
    H = np.zeros((N, N))
    
    for i in range(N):
        for j in range(N):
            S[i, j] = overlap_integral(x_values[i], x_values[j], y_values[i], y_values[j], alpha_x, alpha_y)
            H[i, j] = kinetic_integral(x_values[i], x_values[j], y_values[i], y_values[j], alpha_x, alpha_y) + potential_integral(x_values[i], x_values[j], y_values[i], y_values[j], omega_x, omega_y, alpha_x, alpha_y)
    
    return S, H

# Obliczenie macierzy S i H
S, H = calculate_S_H_matrices(N, alpha_x, alpha_y, w_x, w_y)
```

###  Rozwiązywanie problemu własnego

```python
# Funkcja rozwiązująca problem własny
def solve_eigenvalue_problem(H, S):
    # Ustalenie precyzji obliczeń
    tol = 1e-8
    
    # Obliczenie macierzy odwrotnej do pierwiastka macierzy S
    inv_sqrt_S = np.linalg.inv(np.sqrt(S))
    
    # Przekształcenie macierzy H do postaci w nowej bazie
    H_prime = np.dot(inv_sqrt_S, np.dot(H, inv_sqrt_S))
    
    # Rozwiązanie problemu własnego
    eigenvalues, eigenvectors = eigh(H_prime)
    
    # Upewnienie się, że wartości własne są rzeczywiste i wektory własne są hermitowskie
    eigenvalues = np.real_if_close(eigenvalues, tol=tol)
    eigenvectors = np.real_if_close(eigenvectors, tol=tol)
    
    # Posortowanie wartości własnych i wektorów własnych
    sorted_indices = np.argsort(eigenvalues)
    sorted_eigenvalues = eigenvalues[sorted_indices]
    sorted_eigenvectors = eigenvectors[:, sorted_indices]
    
    return sorted_eigenvalues, sorted_eigenvectors

# Rozwiązanie problemu własnego
eigenvalues, eigenvectors = solve_eigenvalue_problem(H, S)

# Wydrukowanie wartości własnych
print("Wartości własne:")
print(eigenvalues)

# Wydrukowanie wektorów własnych (jako kolumny)
print("\nWektory własne:")
print(eigenvectors)
```

## Zadanie 3: Obliczanie kwadratów modułów funkcji falowej sześciu najniższych stanów

```python
Δx = 1 / 0.05292  # Wartość w atomowych
Δy = Δx  # Zakładamy symetryczną siatkę
a = Δx * (N - 1) / 2

x_values = np.linspace(-a, a, N)
y_values = np.linspace(-a, a, N)

num = 100
Xa = np.linspace(-a, a, num)
Ya = np.linspace(-a, a, num)

# Rozwiązanie problemu własnego
eigenvalues, eigenvectors = solve_eigenvalue_problem(H, S)

# Posortowanie wartości własnych i wektorów własnych
sorted_indices = np.argsort(eigenvalues)
sorted_eigenvalues = eigenvalues[sorted_indices]
sorted_eigenvectors = eigenvectors[:, sorted_indices]

# Wybierz 6 najniższych stanów własnych
lowest_eigenvalues = sorted_eigenvalues[:6]
lowest_eigenvectors = sorted_eigenvectors[:, :6]

# Zbuduj funkcje falowe dla 6 najniższych stanów własnych
psi_lowest_states = np.zeros((num, num, 6))
for k in range(6):  # Iterujemy tylko po 6 najniższych stanach własnych
    for i in range(num):
        for j in range(num):
            for l in range(N):
                psi_lowest_states[i, j, k] += φ(Xa[i], x_values[l], alpha_x) * φ(Ya[j], y_values[l], alpha_y) * lowest_eigenvectors[l, k]

a *= 0.05292           
                
# Narysuj mapy kwadratów modułów funkcji falowych dla 6 najniższych stanów własnych
plt.figure(figsize=(16, 12))
for i in range(6):
    plt.subplot(3, 2, i+1)
    plt.imshow(np.abs(psi_lowest_states[:, :, i])**2, extent=[-a, a, -a, a], origin='lower', cmap='hot')
    plt.colorbar(label=r'$|\psi|^2$')
    plt.title(f'State {i+1}, E = {lowest_eigenvalues[i]:.2f}')
    plt.xlabel('x (nm)')
    plt.ylabel('y (nm)')
plt.tight_layout()
plt.savefig('zadanie_3.png')
plt.show()
```

## Zadanie 4: Wyliczanie energii 10 najniższych stanów w funkcji ωx i rysowanie wykresu E(ωx)

```python
omega_x_values = np.linspace(0, 500 / E_h, 100)

for i in range(10):
    plt.figure(figsize=(8, 6))
    
    lowest_energies = []
    for omega_x in omega_x_values:
        w_x = omega_x
        S, H = calculate_S_H_matrices(N, alpha_x, alpha_y, w_x, w_y)
        eigenvalues, _ = solve_eigenvalue_problem(H, S)
        lowest_energies.append(eigenvalues[i])

    lowest_energies = np.array(lowest_energies)

    plt.plot(omega_x_values * E_h, lowest_energies, label=f'Obliczone dla State {i+1}')

    # Krzywa teoretyczna
    analytical_energy = (i + 0.5) * (hw_x + hw_y) / E_h
    plt.plot(omega_x_values * E_h, np.full_like(omega_x_values, analytical_energy), 'k--', label=f'Krzywa teoretyczna dla State {i+1}')

    plt.xlabel(r'$\omega_x$ (meV)')
    plt.ylabel('E (meV)')
    plt.title(f'Porównanie dla State {i+1}')
    plt.legend()
    plt.grid(True)
    
    plt.savefig(f'wykres_{i+1}.png')
    plt.close()
```
